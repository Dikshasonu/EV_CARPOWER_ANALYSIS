```python
import seaborn as sns
import matplotlib.pyplot as plt
print(data.columns)
sns.displot(data['Power(kW)'], kde=True)  # kde=True adds a kernel density estimate
plt.title('Distribution of Power(kW)')
plt.show()








```

```python
import matplotlib.pyplot as plt
fig,axes = plt.subplots(5,1, figsize=(20, 30))
axi=0
for col in data.select_dtypes( include='float64'):
    sns.countplot(x=data[col],ax=axes[axi])
    axi+=1
for col in data.select_dtypes( include='int64'):
    sns.countplot(x=data[col],ax=axes[axi])
    axi+=1
for col in data.select_dtypes( include='object'):
    sns.countplot(x=data[col],ax=axes[axi])
    axi+=1
           
plt.tight_layout()
```
